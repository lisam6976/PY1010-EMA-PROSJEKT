### I denne prosjektoppgaven √∏nsker jeg √• analysere ankomstene av enslige mindre√•rige asyls√∏kere (EMA) i 2017. Tallene p√• ankomster per m√•ned er hentet fra üîó[UDI](https://www.udi.no/statistikk-og-analyse/statistikk/asylsoknader-enslige-mindrearige-asylsokere-2017/), men kj√∏nn og alder er fiktive.

#### Denne grafen viser fordelingen av EMA-ankomster til Norge i 2017, visualisert i et stolpediagram. 

```python
import numpy as np #fordi jeg skal ha arrays i koden, gj√∏re beregning
import pandas as pd #For √• lese excel-fil
import matplotlib.pyplot as plt #for √• kunne lage diagrammer
data = pd.read_excel("EMA2017.xlsx") #for √• lese excel-filen jeg har laget med data fra UDI og fiktiv data.

maaned = data['Maaned'].values
u_dag = data['ukedag'].values
land = data['land'].values
kjonn = data['kjonn'].values
alder = data['alder'].values
#Her lagrer jeg hver enkelt kolonne fra datasettet som arrays lagret med .values og tilegner navn p√• informasjonen de lagrer.

EMA_mnd = data['Maaned'].value_counts() #beregner antall forekomster av landene ved bruk av Pandas vektoriserte beregnigner.

month_order = ['januar', 'februar', 'mars', 'april', 'mai', 'juni', 'juli', 'august', 'september', 'oktober', 'november', 'desember']
EMA_mnd = EMA_mnd.reindex(month_order, fill_value=0) #For at programmet skal organisere m√•nedene i riktig rekkef√∏lge, ikke i rekkef√∏lge st√∏rst-minst

#Under lager jeg stolpediagrammet med farger, titler og ruterett. 
plt.figure(figsize=(10, 6))
EMA_mnd.plot(kind='bar', color='green', edgecolor='black')

plt.title('Antall EMA per m√•ned i 2017', fontsize=16)
plt.xlabel('') #Jeg √∏nsker ikke en tittel p√• x-aksen fordi det holder med navnene p√• m√•nedene.
plt.ylabel('Antall EMA', fontsize=14)
plt.xticks(rotation=45) #for √• 'tilte' titlene p√• x-aksen, √∏ker lesbarhet
plt.grid(axis='y', linestyle='--', alpha=0.7) #

plt.show() #viser diagrammet.

```

#### N√• √∏nsker jeg √• finne ut hvilke ukedager som har de h√∏yeste ankomstene av EMA. Dette kan v√¶re praktisk √• vite for 1.linje i utlendingsforvaltningen n√•r de skal planlegge n√•r behovet for personell, tolker og representanter er st√∏rst. 

```python
ukedag_antall = data['ukedag'].value_counts()
#jeg lager en egendefinert funksjon for √• lage et kakediagram slik at jeg kan bruke det senere i programmet ogs√•.
def egendefinert_kake(data, tittel, colors=None):
    plt.figure(figsize=(8, 8))
    data.plot(kind='pie', autopct='%1.f%%', startangle=90, colors=colors, shadow=False) #I denne koden ligger utregningen av prosent i 'autopct' og jeg har valgt '1.' for at det ikke skal v√¶re desimaltall fordi det er snakk om mennesker som er 1.
    plt.title(tittel, fontsize=16)
    plt.ylabel('') #for √• ikke ha en y-tittel
    plt.show()


egendefinert_kake(
    data = ukedag_antall,
    tittel = 'S√∏knader fra EMA fordelt p√• ukedager',
    colors=plt.cm.Paired.colors,
)
    


#For √• sjekke at 'autopct' har regnet riktig prosent skriver jeg inn utregningen ogs√•
ukedag_prosent = ((ukedag_antall / ukedag_antall.sum()) * 100).round() #Rundet opp for √• f√• et heltall. Her kunne jeg ogs√• endret typen til 'integer', men vil da f√• kun heltallet, og ikke avrundet til n√¶rmeste heltall. 

#N√•r dette skriver ut √∏nsker jeg √• f√• dagene i riktig rekkef√∏lge.
riktig_rekkef√∏lge = ['mandag', 'tirsdag', 'onsdag', 'torsdag', 'fredag', 'lordag', 'sondag']

ukedag_prosent = ukedag_prosent.reindex(riktig_rekkef√∏lge)

print(ukedag_prosent)


```

### Jeg √∏nsker √• finne ut hvilke kontinenter EMA i 2017 kom fra og hvordan fordelingen er.

```python
file_path = ('EMA2017.xlsx')
data = pd.read_excel(file_path) #henter og leser datasettet fra excel.

kontinenter = ['Afrika', 'Europa', 'Asia', 'Nord Amerika', 'Sor Amerika', 'Australia'] #Legger til kontinentene
landegrupper = {
    'Afrika': ['ERI', 'MAR', 'SOM', 'NGA', 'ETH', 'SDN', 'ALG', 'EGY', 'GIN', 'DRC', 'CIV'],
    'Asia': ['AFG', 'SYR', 'IRQ', 'IRN', 'LBN', 'PHL', 'CHN'],
    'Europa': ['ALB', 'BIH'],
    'Nord Amerika': ['USA'],
    'Sor Amerika': [],
    'Australia': [],
} #Fordeler landene i datasettet til det riktige kontinentet i en dictionary.

kontinent_dictionary = {land: kontinent for kontinent, lands in landegrupper.items() for land in lands} #Lager en dictionary for landene og kontinentene de tilh√∏rer.

data['land_til_kontinent'] = data['land'].map(kontinent_dictionary) #for √• endre fra land til kontinent som fodelt i koden over. 

kontinenter = data['land_til_kontinent'].value_counts() #Teller landene i datasettet.

#Jeg vil ha et diagram som viser fordelingen av kontinenter EMAene kommer fra i prosent.
import matplotlib.pyplot as plt
kontinent_prosent = (kontinenter / kontinenter.sum()) * 100 #regne ut prosent
kontinent_prosent.plot(kind='barh', color=plt.cm.Paired.colors, figsize=(10, 6))
plt.title('Fordelingen av hvilke kontinenter EMA kommer fra', fontsize=16)
plt.ylabel('')
plt.xlabel('Prosentandel') #for √• tydeliggj√∏re at det er prosent som vises.

plt.show()


    
               
    
```

```python
import pandas as pd
# For √• komme tilbake til det originale datasettet og jobbe etter land igjen, ikke kontinent, s√• laster jeg inn datasettet p√• nytt. 
data = pd.read_excel('EMA2017.xlsx')

gjennomsnittsalder_for_land = data.groupby('land')['alder'].mean().round().astype(int) #finner gjennomsnittsalderen for hvert land i et avrundet tall uten desimaler (integer) 
print(gjennomsnittsalder_for_land)


df = gjennomsnittsalder_for_land.reset_index() #lager en dataframe for √• strukturere i kolonner og rader. Resetter indeksen for √• gj√∏re 'land' til en kolonne, ikke indeks. 
df.columns = ['Land', 'Gjennomsnittsalder'] #Navngi kolonnene.

#lage det til en excel-fil
df.to_excel('gjennomsnittsalder_for_land.xlsx', index=False)

print('Gjennomsnittsalder for de ulike landene er n√• skrevet til filen gjennomsnittsalder_for_land.xlsx.')

```

### Her vil jeg at du som leser dette skal skrive inn en tresiffret landekode for s√• f√• til svar hva som var gjennomsnittsalderen p√• EMA fra dette landet i 2017.

```python
import numpy as np

#lagrer informasjon om landene med tilordnet gjennomsnittsalder.

land_numpy = np.array(['AFG', 'ALB', 'ALG', 'BIH', 'CHN', 'CIV', 'DRC', 'EGY', 'ERI', 'ETH', 'GIN', 'IRN', 'IRQ', 'LBN', 'MAR', 'NGA', 'PHL', 'SDN', 'SOM', 'SYR', 'USA','XXX'])

alder_numpy = np.array([14, 17, 17, 17, 16, 17, 16, 16, 15, 16, 17, 15, 14, 17, 16, 17, 16, 16, 15, 13, 17, 16])

print('Her er alle landene og gjennomnittsalderern p√• EMA fra dette landet i 2017')
for i in range(len(land_numpy)):
    print(f'land: {land_numpy[i]}, Gjennomsnittsalder: {alder_numpy[i]} √•r')
    
while True: #Lage en while-l√∏kke for at leseren skal kunne skrive inn flere landekoder for √• se gjennomsnittsalder.
    ditt_land = input('Skriv inn landekoden du vil se gjennomsnittsalderen i 2017 til, n√•r du vil avslutte skriver du -slutt: ') #Lager et program hvor leseren kan legge inn et eget land.
    #hvis landet er lagret i arrayene over:
    if ditt_land.lower() == 'slutt':
        print('Avslutter programmet')
        break
    
    if ditt_land in land_numpy:
        index = np.where(land_numpy == ditt_land)[0][0]
        gjennomsnitt_alder = alder_numpy[index]
        print(f'Gjennomsnittsalder for {ditt_land} er {gjennomsnitt_alder} √•r.')
#Hvis landet som blir skrevet inn ikke er lagret i arrayene over.     
    else:
        print('Det kom ingen EMA fra dette landet i 2017')

```
